```python
import random

import matplotlib.pyplot as plt
# 解决折线图中中文显示问题，下面的代码只需运行一次即可
# 如果是Mac本, 不支持SimHei的时候, 可以修改为 'Microsoft YaHei' 或者 'Arial Unicode MS'
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 1.创建画布_纸
plt.figure(figsize=(120,20), dpi=120)
```

```python
# 2.准备数据
# 2.1准备x轴数据
x = range(60)
# 2.2准备y轴数据
y_data1_beijing = [random.uniform(2,30) for i in x]
y_data1_tianjin = [random.uniform(0,30) for i in x]

# 3.具体绘制动作
plt.plot(x,y_data1_beijing,color='#0000FF',label='北京')
plt.plot(x,y_data1_tianjin,color='#FF0000',label='天津')

# 3.1创建x轴的刻度 以及x轴 y轴的刻度范围
x_lable = [f"11点{i}分" for i in x]
y_lable = range(35)

# 3.2
plt.xticks(x[::5],x_lable[::5])
plt.yticks(y_lable[::5])

# 4.标题  x轴 y轴标签
plt.title("北京、天津一个小时温度折线", fontsize=20)
plt.xlabel('时间',fontsize=12)
plt.ylabel('温度',fontsize=12)

# 5.增加网格效果
plt.grid(linestyle='--',alpha=0.5)

# 6.增加图例
plt.legend(loc='best')

plt.show()

```
