```python
import random
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
# 解决折线图中中文显示问题，下面的代码只需运行一次即可
# 如果是Mac本, 不支持SimHei的时候, 可以修改为 'Microsoft YaHei' 或者 'Arial Unicode MS'
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

os.chdir(r"E:\ProjectCode\PythonProjectCode\pythonProject\HaskMaProject")
```

```python
# 柱状图 bar()

# 1.创建画布
plt.figure(figsize=(30,15),dpi=30)

# 2.造数据
x = ['一级能效','二级能效','三级能效','四级能效','五级能效']
y = [10,30,60,120,300]

# 3.绘制
plt.bar(x,y)
plt.grid(linestyle='-.',alpha=0.5)

plt.title("柱状图",fontsize=20)
plt.xlabel("能效等级",fontsize=20)
plt.ylabel("功耗",fontsize=20)

# 4.展示图形
plt.show()
```

```python
# 直方图   hist()  -> 适用于连续的数据

# 1.创建画布
plt.figure(figsize=(30,20),dpi=80)

# 2.造数据
data = np.random.randn(500000)      # 生成500个随机数

# 3.绘制
plt.hist(data,bins=500)
plt.title("直方图",fontsize=20)

plt.show()


```

```python
# 饼图    pie()      -->    使用比例统计

# 2.造数据
sizes = [20, 10, 50, 20]
datalabels = ['鸡蛋','玉米','豆腐脑（咸）','豆浆（甜）']

# 1.绘制画布
# plt.pie(sizes,labels = datalabels, autopct='%1.1f%%')

explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=datalabels, autopct='%1.1f%%',
    shadow=True, startangle=90)


plt.show()




```

```python
# 4.散点图   scatter  -->   适用于散点数据的统计

# 2.造数据
x = [12,3,4,5]
y = [2,3,4,5]

# 1.绘制画布
plt.scatter(x,y,color='r')
plt.show()

```
